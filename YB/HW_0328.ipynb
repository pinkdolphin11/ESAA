[다변량 그래프 그리기]

다변량 그래프 : 3개 이상의 변수를 사용한 그래프



> 다변량 데이터로 다변량 그래프 그리기 - 산점도 그래프

1. 앞에서 실습한 산점도 그래프에 성별을 새 변수로 추가하기

```python
#Female과 같은 문자열은 색상을 지정하는 값으로 사용할 수 없으므로 각 문자열을 정수로 치환하는 함수를 만들어 사용
def recode_sex(sex):
  if sex=='Female':
    return 0
  else:
    return 1
```

2. record_sex 메서드가 반환한 값(0,1)을 데이터프레임에 sex_color 열을 새로 생성하여 추가

```python
tips['sex_color']=tips['sex'].apply(recode_sex)
```

3. 테이블당 인원 수(size)를 산점도 그래프에 추가

```python
#scatter 메서드에 s(점의 크기, size),c(점의 색상, color) 인자값으로 테이블당 인원 수와 성별의 차환값을 전달
scatter_plot=plt.figure()
axes1=scatter_plot.add_subplot(1,1,1)
axes1.scatter(x=tips['total_bill'],y=tips['tip'],s=tips['size']*10,c=tips['sex_color'],alpha=0.5)
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

> 단변량 그래프 그리기 - 히스토그램



1. tips 데이터 불러오기

```python
import seaborn as sns
tips=sns.load_dataset('tips')
```

2. seaborn 라이브러리로 subplots, distplot 메서드를 사용하여 히스토그램 그리기

```python
ax=plt.subplots() #기본 틀 만들기
ax=sns.distplot(tips['total_bill']) #distplot 메서드에 total_bill 열 데이터 전달
ax.set_title('Total Bill Histogram with Density Plot')
```

3. 밀집도 그래프 제외

* distplot 메서드는 히스토그램과 밀집도 그래프를 같이 그린다. 밀집도 그래프를 제외하고 싶다면 kde 인자값을 False로 설정

```python
ax=plt.subplots()
ax=sns.distplot(tips['total_bill'],kde=False)
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

4. 밀집도 그래프만 나타내기
* hist 인자를 False로 지정

```python
ax=plt.subplots()
ax=sns.distplot(tips['total_bill'],hist=False)
ax.set_title('Total Bill Density')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')
```

5. distplot 메서드에 rug 인자를 추가하여 양탄자(rug) 그래프까지 그리기

6. countplot 메서드에 tips 데이터프레임의 day 열 데이터를 넣어 count 그래프 그리기

```python
ax=plt.subplots()
ax=sns.countplot('day',data=tips)
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```



> 다양한 종류의 이변량 그래프 그리기



1. seaborn 라이브러리로 산점도 그래프 그리기
* seaborn 라이브러리의 regplot 메서드를 사용하면 산점도 그래프와 회귀선을 함께 그림

* fit_reg 인자를 False로 지정하면 회귀선 제거

```python
ax=plt.subplots()
ax=sns.regplot(x='total_bill',y='tip',data=tips)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax=plt.subplots()
ax=sns.regplot(x='total_bill',y='tip',data=tips,fit_reg=False)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

2. 산점도 그래프와 히스토그램을 한 번에 그리는 jointplot 메서드

```python
joint=sns.jointplot(x='total_bill',y='tip',data=tips)
joint.set_axis_labels(xlabel='Total Bill',ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip',fontsize=10,y=1.03)
```

3. 육각 그래프(hexbin) : 2차원 표면에 육각형으로 데이터를 쌓아 표현
* 특정 데이터의 개수가 많아지면 점점 진한 색으로 표현

* jointplot 메서드를 그대로 사용하고, kind 인자값을 hex로 지정

```python
hexbin=sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
hexbin.set_axis_labels(xlabel='Total Bill',ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip',fontsize=10,y=1.03)
```

4.  kdeplot 메서드로 이차원 밀집도 그리기

```python
ax=plt.subplots()
ax=sns.kdeplot(data=tips['total_bill'],data2=tips['tip'],shade=True) #shade 인자를 True로 지정하여 음영 효과 주기
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

5. 지정한 변수의 평균을 계산하여 바 그래프 그리기

* 관계 그래프의 절반(왼쪽 위에서 오른쪽 아래로 그린 대각선 기준 위)은 나머지 절반과 같다.

* map_upper/lower 메서드 : 대각선을 기준으로 위쪽/아래쪽에 그릴 그래프 지정
* map_diag 메서드 : 대각선을 중심으로 그래프 그림

```python
pair_grid=sns.PairGrid(tips)
pair_grid=pair_grid.map_upper(sns.regplot)
pair_grid=pair_grid.map_lower(sns.kdeplot)
pair_grid=pair_grid.map_diag(sns.distplot,rug=True)
plt.show()
```

2. whitegrid로 스타일을 설정하여 그래프 그리기

```python
sns.set_style('whitegrid') #그래프에 가로줄이 생김
fig, ax=plt.subplots()
ax=sns.violinplot(x='time',y='total_bill',hue='sex',data=tips,split=True)
```

3. for문을 이용하여 모든 스타일을 하나씩 적용한 그래프 그리기

```python
fig=plt.figure()

seaborn_styles=['darkgrid','whitegrid','dark','white','ticks']

for idx, style in enumerate(seaborn_styles):
  plot_position = idx+1
  with sns.axes_style(style):
    ax = fig.add_subplot(2,3,plot_position)
    violin = sns.violinplot(x='time',y='total_bill',data=tips, ax=ax)
    violin.set_title(style)

fig.tight_layout()
```
